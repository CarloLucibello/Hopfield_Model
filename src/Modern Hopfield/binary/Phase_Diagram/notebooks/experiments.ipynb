# Histograms

```julia
t = 100
m = zeros(t)
N = 50
α = 0.06
λ = 0.05
M = round(Int, exp(N * α))

for i in 1:t
    ξ = MHB.generate_patterns(M, N)
    k = rand(1:M)
    σ = ξ[:, k]
    σ_rec = MHB.monte_carlo(σ, ξ; λ = λ, β = 100)
    m[i] = MHB.overlap(σ_rec, σ)
end
```

```julia
histogram(m, nbins = 30)
```
